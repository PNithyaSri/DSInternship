### Data Cleaning

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
def get_third_element(x):
    elements = x.split('\t')
    if len(elements) >= 3:
        return elements[2]
    else:
        return None  

data["Text"] = data["Text"].apply(get_third_element)
```

```python
data["Text"] = data["Text"].astype(str)
```

### How many LinkedIn URLs were there over all the chat files?

### Out of all the LinkedIn URLs shared over chat, how many are valid profile URLs?
